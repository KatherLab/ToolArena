```python
from toolarena import ToolDefinition
from toolarena.utils import DATA_DIR, DEFINITIONS_DIR
```

```python
task_dir = DEFINITIONS_DIR / "conch_extract_features"
definition = ToolDefinition.from_yaml(task_dir / "task.yaml")
definition
```

```python
implementation = definition.build(
    install_script=task_dir.joinpath("install.sh").read_text(),
    code_implementation=task_dir.joinpath("implementation.py").read_text(),
)
implementation
```

```python
result = implementation.run(
    invocation=definition.example, data_dir=DATA_DIR / definition.name / "data"
)
result
```
